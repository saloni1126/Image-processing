```python
import numpy as np
import matplotlib.pyplot as plt

def rectangular_pulse(t, width):
    return np.where((t>0) & (t < width) , 1, 0)

def exponential_decay(t, decay_constant):
    return np.exp(-decay_constant * t) * (t >= 0)

def convolution(signal1, signal2, dt):
    result = np.convolve(signal1, signal2, mode='full') * dt
    t_result = np.linspace(0, (len(result) - 1) * dt, len(result))
    return t_result, result

# Parameters
dt = 0.01  # Time step
t = np.arange(-10, 20, dt)  # Time range

# Create signals
width = 4
decay_constant = 0.5
rect_pulse = rectangular_pulse(t, width)
exp_decay = exponential_decay(t, decay_constant)

# Perform convolution
t_conv, conv_result = convolution(rect_pulse, exp_decay, dt)

# Plotting
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(t, rect_pulse, label='Rectangular Pulse')
plt.legend()
plt.grid()

plt.subplot(3, 1, 2)
plt.plot(t, exp_decay, label='Exponential Decay')
plt.legend()
plt.grid()

plt.subplot(3, 1, 3)
plt.plot(t_conv, conv_result, label='Convolution')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
