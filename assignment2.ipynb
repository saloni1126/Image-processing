```python
import numpy as np

# Set the seed for reproducibility (optional)
np.random.seed(42)

#Generate uniform values
uniform_values = np.random.uniform(low=0.0, high=1.0, size=100000)

#Print the values
print(uniform_values)
```

```python
# Uniform to traingle
y = 1- np.sqrt(1 - uniform_values)
np.append(y,1+ np.sqrt(1 - uniform_values))
```

```python
import matplotlib.pyplot as plt
plt.hist(y,bins=150)
plt.show()
```

```python
z = np.sqrt(2*y - y**2)
np.append(z,-np.sqrt(2*y - y**2))
```

```python
plt.hist(z,bins=150)
plt.show()
```
