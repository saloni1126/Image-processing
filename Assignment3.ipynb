Implement Edge detection using

(I) DoG Filter

(ii) LoG Filter and

(iii) Canny Edge Detector and compare the results

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import urllib.request

# Load the image from a URL
url = 'https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcRUhzCK4hOHwXpRAP2NKYj3BZEDgqHEaXTgsQ&usqp=CAU'
with urllib.request.urlopen(url) as url:
    s = url.read()
arr = np.asarray(bytearray(s), dtype=np.uint8)
image = cv2.imdecode(arr, -1) # 'Load it as grayscale'

# DoG Filter
gaussian_blur1 = cv2.GaussianBlur(image, (5,5), 0)
gaussian_blur2 = cv2.GaussianBlur(image, (9,9), 0)
dog = gaussian_blur1 - gaussian_blur2

# LoG Filter
log = cv2.Laplacian(gaussian_blur1, cv2.CV_64F)

# Canny Edge Detector
canny = cv2.Canny(image, 100, 200)

# Display the results
titles = ['Original Image', 'DoG Filter', 'LoG Filter', 'Canny Edge Detector']
images = [image, dog, log, canny]

for i in range(4):
    plt.subplot(2, 2, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()
```
